```python

    
def generate_chromosome ():
    ale = random.sample(allels, len(allels))
  
    # Create an empty 2D array to represent the timetable
    timetable = np.empty((num_rooms, num_time_slots), dtype='U100')

    # Fill the timetable with the course, section, and teacher information from the allels list
    for ale in ale:
        course, section, teacher = ale
        for i in range(num_time_slots):
            # Find a free room for the current time slot
            free_rooms = np.where(timetable[:, i] == '')[0]
            if len(free_rooms) > 0:
                room_index = np.random.choice(free_rooms)
                timetable[room_index, i] = f'{course} ({section}) - {teacher}'
                break
    return timetable



def fitness_function(timetable):
    # Convert the timetable to a pandas DataFrame
    df = pd.DataFrame(timetable)
    
    # Initialize the fitness score to 0
    fitness_score = 0
    flag1 = 1
    flag2 = 1
    
    # Check for clashes within each column (i.e., each time slot)
    for col in df.columns:
        # Extract the section codes from each cell in the column
        section_codes = df[col].apply(lambda x: x.split('(')[-1].split(')')[0])
#         print(section_codes)
        section_count = section_codes.value_counts()
#         print(section_count)
        if section_count.max() > 1:
            flag2 = 0 
            break
        
        # Extract the teacher names from each cell in the column
        teacher_names = df[col].apply(lambda x: x.split(' - ')[-1])
#         print(teacher_names)
        teacher_count = teacher_names.value_counts()
#         print(teacher_count)
        if teacher_count.max() > 1:
            flag1 = 0
            break
    
    # Add fitness score based on flag values
    if flag1:
        fitness_score += 35
    if flag2:
        fitness_score += 35 
        
    # Check for section and teacher frequency across all columns
    section_codes = df.apply(lambda x: x.str.split('(').str[-1].str.split(')').str[0])
    section_count = section_codes.apply(pd.Series.value_counts).fillna(0)
    if (section_count > 3).sum().sum() > 0:
        fitness_score += 0
    else:
        fitness_score += 15
    
    teacher_names = df.apply(lambda x: x.str.split(' - ').str[-1])
    teacher_count = teacher_names.apply(pd.Series.value_counts).fillna(0)
    if (teacher_count > 3).sum().sum() > 0:
        fitness_score += 0
    else:
        fitness_score += 15
    
    return fitness_score


    
def tournament_selection(population, tournament_size):
#     print(len(population))
    random_integers = np.random.randint(0, len(population), tournament_size)
#     print(random_integers)
    population_dict = {}
    
    for i in range(tournament_size):
        chromosome = tuple(population[random_integers[i]].flatten())
        fitness = fitness_function(population[random_integers[i]])
        population_dict[chromosome] = fitness
    
    # Sort the population_dict based on fitness scores
    sorted_population = sorted(population_dict.items(), key=lambda x: x[1], reverse=True)
    parent1 = np.array(sorted_population[0][0]).reshape(num_rooms, num_time_slots)
    parent2 = np.array(sorted_population[1][0]).reshape(num_rooms, num_time_slots)
    
    
    return parent1, parent2


def crossover(parent1, parent2):
    child1 = np.empty((num_rooms, num_time_slots), dtype='U100')
    child2 = np.empty((num_rooms, num_time_slots), dtype='U100')

    # Copy first 3 columns of parent 1 to child 1
    child1[:, :3] = parent1[:, :3]

    # Copy last 3 columns of parent 2 to child 1
    child1[:, 3:] = parent2[:, 3:]

    # Copy first 3 columns of parent 2 to child 2
    child2[:, :3] = parent2[:, :3]

    # Copy last 3 columns of parent 1 to child 2
    child2[:, 3:] = parent1[:, 3:]
    return child1,child2

def mutation(individual, probability):
    mutated_individual = individual.copy()
    if np.random.uniform() < probability:
        i = np.random.randint(num_rooms)
        j = np.random.randint(num_time_slots)
        new_allele = allels[np.random.randint(len(allels))]
        new_course, new_section, new_teacher = new_allele
        mutated_individual[i][j] = f'{new_course} ({new_section}) - {new_teacher}'
    return mutated_individual


generations = 30000
population_size = 50
for generation in range(generations):
    print(f"Generation {generation+1}")
    # Selection
    new_population = []
    for i in range(population_size//2):
        a= tournament_selection(population,10)
        b=crossover(a[0], a[1])
        child1 = b[0]
        child2 = b[1]
        if generation %50==0:
            child1 = mutation(b[0], 0.9)
            child2 = mutation(b[1], 0.9) 
        new_population.append(child1)
        new_population.append(child2)
    new_population[0]=(max(population, key=fitness_function))
    population = new_population
    best_individual = max(population, key=fitness_function)
    print(f"Best individual: {fitness_function(best_individual)}")
    if fitness_function(best_individual)==100:
        print("Solution found!")
        break
```

```python
num_rooms = 35
num_time_slots = 6

class Course:
    def __init__(self, code, teacher):
        self.code = code
        self.teacher = teacher

def generate_chromosome():
    ale = random.sample(allels, len(allels))
    timetable = np.empty((num_rooms, num_time_slots), dtype=object)
    for ale in ale:
        course, teacher = ale
        course_obj = Course(course, teacher)
        for i in range(num_time_slots):
            free_rooms = np.where(timetable[:, i] == None)[0]
            if len(free_rooms) > 0:
                room_index = np.random.choice(free_rooms)
                timetable[room_index, i] = course_obj
                break
    return timetable

def fitness_function(timetable):
    section = set(item.code for sublist in timetable for item in sublist)
    s = {sec: 0 for sec in section}
    teacher = set(item.teacher for sublist in timetable for item in sublist)
    t = {tea: 0 for tea in teacher}
    all_less_than_3_t = True
    all_less_than_3_s = True
    fitness_score = 0
    teacher_names = []
    section_codes = []
    check=True
    check1=True
    for i in range(6):
        for j in range(len(timetable)):
            if timetable[j][i].code not in section_codes:
                section_codes.append(timetable[j][i].code)
                flag1 = True
            else:
                flag1 = False
                check=False
                
            if timetable[j][i].teacher not in teacher_names:
                teacher_names.append(timetable[j][i].teacher)
                flag2 = True
            else:
                flag2 = False
                check1=False
            s[timetable[j][i].code] += 1
            t[timetable[j][i].teacher] += 1

    if flag1 :
        fitness_score += 35
    if flag2:
        fitness_score += 35

    for count in t.values():
        if count > 4:
            all_less_than_3_t = False
            break
    for count in s.values():
        if count > 4:
            all_less_than_3_s = False
            break
    if all_less_than_3_t:
        fitness_score += 15
    if all_less_than_3_s:
        fitness_score += 15

    return fitness_score

def crossover(parent1, parent2):
    child1 = np.empty((num_rooms, num_time_slots), dtype=object)
    child2 = np.empty((num_rooms, num_time_slots), dtype=object)
    child1[:16, :] = parent1[:16, :]
    child1[16:, :] = parent2[16:, :]
    child2[:16, :] = parent2[:16, :]
    child2[16:, :] = parent1[16:, :]
    child1[:, :3] = parent1[:, :3]
    child1[:, 3:] = parent2[:, 3:]
    child2[:, :3] = parent2[:, :3]
    child2[:, 3:] = parent1[:, 3:]
    return child1, child2

def mutation(individual, probability):
    mutated_individual = individual.copy()
    i = np.random.randint(num_rooms)
    j = np.random.randint(num_time_slots)

    # Get the current value in the chosen slot
    current_value = mutated_individual[i][j]

    # Replace the current value with a random non-empty value
    alleles = []
    for row in mutated_individual:
        for value in row:
            if value is not None:
                alleles.append((value.code, value.teacher))
    non_empty_alleles = list(set(alleles))
    new_ale = random.choice(non_empty_alleles)
    new_value = Course(new_ale[0], new_ale[1])

    mutated_individual[i][j] = new_value

    return mutated_individual


  
population_size = 13

import random
import numpy as np
population=[]
for i in range (population_size):
    temp=generate_chromosome()
    population.append(temp)
#     print(fitness_function(temp))
    


generations = 30000

def sel(d):
    c=2
    return random.sample(d,c)
for generation in range(generations):
    print(f"Generation {generation+1}")
    # Selection
    new_population = []
    for i in range(population_size//2):
        a= sel(population)
        b=crossover(a[0], a[1])
        child1 = b[0]
        child2 = b[1]
        if generation%100==0:
            child1 = mutation(b[0],1)
            child2 = mutation(b[1], 1) 
        new_population.append(child1)
        new_population.append(child2)

    new_population[0]=max(population, key=fitness_function)
    population = new_population
    best_individual = max(population, key=fitness_function)
    print(f"Best individual: {fitness_function(best_individual)}")
    if fitness_function(best_individual)==100:
        print("Solution found!")
        break
        
import pandas as pd

# create an empty dictionary to store the data
data = {}

# loop over each time slot
for j in range(num_time_slots):
    # create a list to store the data for this time slot
    slot_data = []
    # loop over each room
    for i in range(num_rooms):
        # add the course code and teacher for this room and time slot
        if best_individual[i][j] is not None:
            slot_data.append(f"{best_individual[i][j].code} ({best_individual[i][j].teacher})")
        else:
            slot_data.append(None)
    # add the list of data for this time slot to the dictionary
    data[f"Time Slot {j+1}"] = slot_data

# create the dataframe from the dictionary
df = pd.DataFrame(data)
df
```
